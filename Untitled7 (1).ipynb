# Import the packages

```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

# download file wav music

```python
z, w = librosa.load('file_example_WAV_1MG.wav') #librosa.util.example_audio_file()
```

# chroma_stft 

```python
chroma_stft = librosa.feature.chroma_stft(y=z, sr=w)
```

```python
#the plot of chroma_stft
plt.figure()
plt.subplot(2,1,1)
librosa.display.specshow(chroma_stft, y_axis='chroma')
plt.title('the plot of chroma_stft')
plt.colorbar()
```

# chroma_cq

```python
chroma_cq = librosa.feature.chroma_cqt(y=z, sr=w)
```

```python
#the plot of chroma_cq
plt.subplot(2,1,2)
librosa.display.specshow(chroma_cq, y_axis='chroma', x_axis='time')
plt.title('the plot of chroma_cq')
plt.colorbar()
```
